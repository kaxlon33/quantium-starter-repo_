```python
import pandas as pd

files = [
    "daily_sales_data_0.csv",
    "daily_sales_data_1.csv",
    "daily_sales_data_2.csv"
]
all_data = []

for file in files:
    df = pd.read_csv(file)
    all_data.append(df)
    
final_df = pd.concat(all_data, ignore_index=True)
```

```python
print(final_df.dtypes)
print(final_df.head())
```

```python
final_df['price'] = (
    final_df['price'].astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
)

final_df['price'] = pd.to_numeric(final_df['price'], errors='coerce')
```

```python
final_df = final_df[final_df['product'].str.strip().str.lower() == 'pink morsel'].copy()
```

```python
final_df['Sales'] = final_df['quantity'] * final_df['price']
```

```python
final_df = final_df[['Sales','date','region']]
```

```python
final_df.to_csv("combined_sales_data.csv", index=False)
```
